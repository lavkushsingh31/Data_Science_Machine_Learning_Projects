# Project - Face Detection using OpenCV's Haarcascades Classifier

```python
import cv2
import matplotlib.pyplot as plt
```

```python
kids_img = cv2.imread("Datasets/rural_school_kids.jpg")

# since opencv reads the image and BGR instead of RBG, we need to fix the order of the array channels
kids_img = cv2.cvtColor(kids_img, cv2.COLOR_BGR2RGB)
```

```python
kids_img[:2, :, :]
```

```python
plt.imshow(kids_img)
plt.axis('off')
plt.show()
```

### Refer the site
- https://github.com/opencv/opencv/tree/4.x/data/haarcascades
    * haarcascade_frontalface_alt.xml
    * haarcascade_frontalface_alt2.xml
    * haarcascade_frontalface_alt_tree.xml

```python
model_frontalface_alt = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
faces = model_frontalface_alt.detectMultiScale(kids_img, scaleFactor= 1.2, minNeighbors = 5 )
```

```python
faces
```

```python
for f in faces:
    x,y,w,h = f
    cv2.rectangle(kids_img, (x,y), (x+w, y+h), (255,0,0), 2 )
```

```python
plt.imshow(kids_img)
plt.axis('off')
plt.show()
```
